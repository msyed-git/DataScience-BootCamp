```python
from sqlalchemy import create_engine 
import pandas as pd

engine = create_engine('sqlite:///sqlite_db_pythonsqlite.db')
```

```python
with engine.connect() as con, con.begin():  
    Bookings = pd.read_sql_table("Bookings", con) 
    
Bookings
```

```python
with engine.connect() as con, con.begin():  
    Facilities = pd.read_sql_table("Facilities", con)
Facilities
```

```python
with engine.connect() as con, con.begin():  
    Members = pd.read_sql_table("Members", con)
Members
```

```python
Q10 = pd.read_sql_query("SELECT name, sum(case when memid = 0 then slots * guestcost else slots * membercost end) as Revenue FROM Bookings LEFT JOIN Facilities using(facid) GROUP BY name HAVING Revenue < 1000",engine)
# list of facilities with a total revenue less than 1000
Q10
```

```python
Q11 = pd.read_sql_query("SELECT m.membername, mm.firstname || ' ' || mm.surname AS Recommended_BY FROM (SELECT firstname || ' ' || surname AS membername, recommendedby FROM Members WHERE recommendedby > 0 ORDER BY surname ASC, firstname ASC) AS m INNER JOIN Members as mm ON m.recommendedby = mm.memid ORDER BY mm.surname ASC, mm.firstname ASC", engine)
#Produce a report of members and who recommended them in alphabetic surname,firstname order
Q11
```

```python
Q12 = pd.read_sql_query("SELECT f.name, m.firstname || ' ' || m.surname AS membername,   COUNT(*) AS usedbymember FROM Bookings AS b INNER JOIN Facilities AS f USING(facid) INNER JOIN members AS m USING(memid) WHERE b.memid != 0 GROUP by f.name, membername ",engine)
##Find the facilities with their usage by member, but not guests 
Q12
```

```python
Q13 = pd.read_sql_query("SELECT f.name, strftime('%m', starttime) as month,   COUNT(*) AS monthlyusage FROM Bookings AS b INNER JOIN Facilities AS f USING(facid) WHERE b.memid != 0 GROUP by month, f.name ",engine)
#Q13: Find the facilities usage by month, but not guests 
Q13
```
